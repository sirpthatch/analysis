```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
WEATHER_V1_DATA = "../data/weather_data.csv"
WEATHER_V2_DATA = "../data/weather_data_v2.csv"

CITY_STATE_RUNNER_COUNTS = "../data/city_state_runner_counts.csv"

weather_v1_df = pd.read_csv(WEATHER_V1_DATA)[["city","state"]]
weather_v2_df = pd.read_csv(WEATHER_V2_DATA)[["city","state"]]

weather_v1_cities = set(zip(weather_v1_df["city"], weather_v1_df["state"]))
weather_v2_cities = set(zip(weather_v2_df["city"], weather_v2_df["state"]))
del weather_v1_df, weather_v2_df
```

```python
def clean_str(val:str) -> str:
    return val.lower().replace("\"","").strip()

city_state_counts = pd.read_csv(CITY_STATE_RUNNER_COUNTS)
city_state_counts["city"] = city_state_counts["city"].apply(clean_str)
city_state_counts["state"] = city_state_counts["state"].apply(clean_str)

display(city_state_counts)
norm_df = pd.DataFrame(city_state_counts.groupby(["city","state"])["runner_count"].sum().reset_index().sort_values(by="runner_count", ascending=False))
norm_df.to_csv("../data/city_state_runner_counts_v2.csv")
```

```python
weather_v1_runners = norm_df[
    norm_df.apply(
        lambda row: (row['city'], row['state']) in weather_v1_cities, 
        axis=1
    )
]["runner_count"].sum()

print(f"Weather V1: {weather_v1_runners:,}")

weather_v2_runners = norm_df[
    norm_df.apply(
        lambda row: (row['city'], row['state']) in weather_v2_cities, 
        axis=1
    )
]["runner_count"].sum()

print(f"Weather V2: {weather_v2_runners:,}")
```
